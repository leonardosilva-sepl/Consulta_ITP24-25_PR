```python
def scrape_url(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        if pd.isna(url) or url == '':
            print("Erro: URL vazia ou não localizada para raspagem.")
            return False, "URL_EMPTY", "URL vazia ou não localizada para raspagem."
        else:
            response = requests.get(url, headers=headers, timeout=20)
            time.sleep(10)
            response.raise_for_status()
            time.sleep(10)
            soup = BeautifulSoup(response.content, 'html.parser')
            return soup, False, False

    except requests.exceptions.HTTPError as http_err:
        error_msg = f"HTTP error occurred for {url}: {http_err}"
        print(error_msg)
        error_cod = http_err.response.status_code
        error_msg = http_err.response.reason
        return False, error_cod, error_msg

    except requests.exceptions.ConnectionError as conn_err:
        error_msg = f"Connection error occurred for {url}: {conn_err}"
        print(error_msg)
        error_cod = type(conn_err).__name__
        error_msg = "ConnectionError"
        return False, error_cod, error_msg

    except requests.exceptions.Timeout as timeout_err:
        error_msg = f"Connection error occurred for {url}: {timeout_err}"
        print(error_msg)
        error_cod = type(timeout_err).__name__
        error_msg = "timeout_err"
        return False, error_cod, error_msg

    except requests.exceptions.RequestException as req_err:
        error_msg = f"An unexpected request error occurred for {url}: {req_err}"
        print(error_msg)
        error_cod = type(req_err).__name__
        error_msg = "req_err"
        return False, error_cod, error_msg

    except Exception as e:
        error_msg = f"An unexpected error occurred for {url}: {e}"
        print(error_msg)
        error_cod = e.response.status_code if hasattr(e, 'response') and hasattr(e.response, 'status_code') else type(e).__name__
        error_msg = e.response.reason if hasattr(e, 'response') and hasattr(e.response, 'reason') else str(e)
        return False, error_cod, error_msg
```

```python
def scrape_info_stealth(driver, url):
    if not url:
        print("Erro: URL vazia ou não localizada para raspagem.")
        return False, "URL_EMPTY", "URL vazia ou não localizada para raspagem."

    try:
        driver.get(url)
        driver.implicitly_wait(20)
        response_text = driver.page_source
        soup = BeautifulSoup(response_text, "html.parser")
        return soup, False, False

    except Exception as e:
        err = f"Erro ao raspar {url}: {e}"
        print(err)
        return False, type(e).__name__, str(e)
```

```python
def localiza_clicks(driver, parametro_navegacao):
    # LOCALIZA SE HÁ ELEMENTOS COM TEXTOS SELECIONADOS
    lista_elementos = []
    vistos = set()

    for p in parametro_navegacao:
      try:
        els_texto = driver.find_elements(By.XPATH, f"//*[contains(text(),'{p}')]") #BUSCA ELEMENTOS QUE TENHAM TEXTO P
        if not els_texto:
          els_texto = []
      except Exception:
        els_texto = []
        continue

      try:
        els_atribs = driver.find_elements(By.XPATH, f"//*[@*[contains(.,'{p}')]]") #BUSCA ELEMENTOS QUE TENHAM ATRIBUTO P
        if not els_atribs:
          els_atribs = []
      except Exception:
        els_atribs = []
        continue

      els = els_texto + els_atribs
      for el in els:
        try:
          txt = el.text
          html = el.get_attribute("outerHTML")
          id = (txt, html)

          if id not in vistos and el not in lista_elementos:
            lista_elementos.append(el)
            vistos.add(id)
        except:
            continue

    return lista_elementos
```

```python
def clicker(driver, el):
    try:
        main_window = driver.current_window_handle
        el.click()
        time.sleep(2)

        if len(driver.window_handles) > 1:
            for handle in driver.window_handles:
                if handle != main_window:
                    driver.switch_to.window(handle)
                    break

    except (ElementClickInterceptedException, ElementNotInteractableException):
        driver.execute_script('arguments[0].click();', el)
        time.sleep(2)
        if len(driver.window_handles) > 1:
            for handle in driver.window_handles:
                if handle != driver.current_window_handle:
                    driver.switch_to.window(handle)
                    break
```

```python
#retorna todo texto para os elementos desejados
def busca_todo_texto(driver, local):
    #LOCALIZACAO
    elements = driver.find_elements(By.CSS_SELECTOR, local)

    #lista de texto
    lista_text = []
    lista_html = []
    for e in elements:
      try:
        if e.text is not None:
          lista_text.append(e.text)
          lista_html.append(e.get_attribute("outerHTML"))

      except Exception as e:
        print(f"Erro ao obter atributos do elemento: {e}")
        pass

    lista_text = list(set(lista_text))
    lista_html = list(set(lista_html))
    return lista_text, lista_html
```

```python
#retorna link para uma parte de texto do link desejado
def busca_link(driver, parametro):
    #FUNCAO DE LOCALIZACAO
    #driver: driver seleneium ativo
    #parametro: lista de parametro string dos atributos que serão buscados, ex.: "radar", "horario"

    lista_elementos = [] #elementos que contenham algum p
    lista = []
    link = []
    html = []

    for p in parametro:
      elementos = driver.find_elements(By.XPATH, f"//*[@*[contains(.,'{p}')]]") #PARAMETRO DE BUSCA EM ATRIBUTOS

      if elementos is not None:
        lista_elementos.extend(elementos)

    for el in lista_elementos:
      if el.tag_name != 'input' and el.tag_name != 'option' and el.tag_name != 'form' and el.tag_name != "select" and el.get_attribute("option") == None:
        try:
          atribs = busca_todo_atributos(driver,el)

          for nome_atributo, valor_atributo in atribs.items():
            if (nome_atributo is not None) and (valor_atributo is not None) and any(p in valor_atributo for p in parametro):
              link.append(el.get_attribute(nome_atributo))
              html.append(el.get_attribute("outerHTML"))
              lista.append(el)

        except Exception as e:
          print(f"Erro ao obter atributos do elemento: {e}")

    lista = list(set(lista))
    link = list(set(link))
    html = list(set(html))

    return link, html, lista
```

```python
def buscador(driver, tipo_busca, busca):
    match tipo_busca:
      case "texto":
            elemento, html_elemento = busca_todo_texto(driver, busca)

      case "link":
            elemento, html_elemento, lista = busca_link(driver, busca)

      case "texto_simples":
            elemento, html_elemento, lista = busca_texto(driver, busca)

    return elemento, html_elemento, lista
```

```python
def scrap_selenium(url, tipo_busca, busca, parametro_navegacao):

  inicio = time.time()
  parametro_navegacao = padroniza_parametro(parametro_navegacao)
  html_selenium = None
  elemento = None
  html_elemento = None
  error_cod = False
  error_msg = False
  texto_click = []
  html_click = []
  elemento_f = None
  html_elemento_f = None
  duracao = 0

  if pd.isna(url) or (isinstance(url, str) and not url.strip()):
    print(f"Erro: URL inválida recebida: {repr(url)}")
    return None, None, None, "URL_INVALID", "URL vazia, None ou NaN.", None, None, None

  if ("https://radardatransparencia.atricon" in url) or ("https://radar.tce.mt." in url) or ("https://www.tce." in url):
    return None, None, None, "URL_VERIFY", "Foi verificado que a URL fornecida se refere diretamente ao Radar Transparencia ou TCE-MT", None, None, None

  driver = None

  try:
    driver = abrir_driver()
    driver.set_page_load_timeout(500)
    driver.set_script_timeout(500)
    driver.get(url)
    wait = WebDriverWait(driver, 60)

    try: ##Parâmetros HTML para busca inicial
        wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
        html_selenium = driver.page_source

    except TimeoutException as e:
        html_selenium = driver.page_source
        error_cod = type(e).__name__
        error_msg = str(e)
        print(f"Timeout esperando carregamento principal para {url}, seguindo mesmo assim.")

    try:
        if verifica_bloqueio(driver):
          error_cod = "URL_BLOCKED"
          error_msg = "URL bloqueada"

        else:
          url_inicial = driver.current_url
          driver.implicitly_wait(10)
          elemento, html_elemento, lista_elementos = buscador(driver,tipo_busca, busca) ##BUSCA DUPLICADA DEVIDO A ERROS DE ASINC
          elemento, html_elemento, lista_elementos = buscador(driver,tipo_busca, busca)


          ##BUSCA SE NÃO HOUVER ELEMENTO CONTENDO PARAMENTRO DE VERIFICACAO
          if not elemento:
            try:

              bt_aceito(driver) ##ACEITAR TERMOS DE USO QUE POSSA BLOQUEAR A TELA DO DRIVER
              lista_elementos = localiza_clicks(driver, parametro_navegacao)

              ##TENTA NAVEGAR A PARTIR DOS CLICKAVEIS LOCALIZADOS
              if lista_elementos:
                for el in lista_elementos:
                  try:
                    texto_click.append(el.text)
                    html_click.append(el.get_attribute("outerHTML"))
                    guia_original = driver.current_window_handle

                    clicker(driver,el)
                    url_click = driver.current_url

                    if driver.windowns_handles > 1 and driver.current_window_handle == guia_original:
                      driver.switch_to.window(driver.windowns_handles[1])

                    elemento, html_elemento, lista_elementos = buscador(driver,tipo_busca, busca)

                    ## RETORNAR A PAGINA INICIAL
                    if not lista_elementos and url_click == url_inicial:
                      pass

                    if not lista_elementos and url_click != url_inicial:
                      if driver.windowns_handles > 1 and driver.current_window_handle != guia_original:
                        driver.close()
                        driver.switch_to.window(guia_original)
                      else:
                        driver.get(url)
                      pass

                    if lista_elementos:
                      elemento = elemento
                      html_elemento = html_elemento
                      break
                    ## RETORNAR A PAGINA INICIAL

                  except:
                    continue


              if not lista_elementos:
                error_cod = "NENHUM_ELEMENTO_NAVEGACAO"
                error_msg = f"Nenhum elemento de interação contendo: {parametro_navegacao}"

            except NoSuchElementException as e:
                error_cod = "NENHUM_ELEMENTO_LOCALIZADO"
                error_msg = f"Nenhum elemento final localizado na página contendo: {busca} ou de interação contendo: {parametro_navegacao}"

    except Exception as e:
        if not error_cod:
            error_cod = type(e).__name__
            error_msg = str(e)
        print(f"Erro na busca de elementos para {url}: {e}")

  except WebDriverException as e:
      error_cod = type(e).__name__
      error_msg = str(e)
      print(f"WebDriver error for {url}: {error_msg}")

  except Exception as e:
      error_cod = type(e).__name__
      error_msg = str(e)
      print(f"General error for {url}: {error_msg}")

  finally:
    if (not html_elemento) and (not error_cod):
      error_cod = "NENHUM_ELEMENTO_LOCALIZADO"
      error_msg = f"Nenhum elemento final localizado na página contendo: {busca} ou de interação contendo: {parametro_navegacao}"

    duracao = round(time.time() - inicio, 2)

    if driver:
      print(f"Tempo de execução para {url}: {duracao}s")
      driver.quit()

  return html_selenium, elemento, html_elemento, error_cod, error_msg, texto_click, html_click, duracao
```

```python
def scrap_retry(url, tipo_busca, busca, parametro_navegacao):
    inicio = time.time()
    parametro_navegacao = padroniza_parametro(parametro_navegacao)
    html_selenium, elemento, html_elemento = None, None, None
    error_cod, error_msg = False, False
    texto_click, html_click = [], []

    if pd.isna(url) or (isinstance(url, str) and not url.strip()):
        return None, None, None, "URL_INVALID", "URL vazia.", None, None, 0

    driver = None
    try:
        # Adicionando jitter aleatório antes de iniciar o driver
        time.sleep(np.random.uniform(2, 5))

        driver = abrir_driver()
        # Timeouts estendidos para servidores municipais lentos
        driver.set_page_load_timeout(240)
        driver.set_script_timeout(240)

        driver.get(url)
        wait = WebDriverWait(driver, 100)

        try:
            wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
            html_selenium = driver.page_source
        except TimeoutException:
            html_selenium = driver.page_source
            error_cod = "TimeoutException"
            error_msg = "Timeout persistente no reprocessamento"

        if not error_cod:
            if verifica_bloqueio(driver):
                error_cod = "URL_BLOCKED"
            else:
                elemento, html_elemento, _ = buscador(driver, tipo_busca, busca)

                if not elemento:
                    bt_aceito(driver)
                    lista_interaction = localiza_clicks(driver, parametro_navegacao)
                    for el in lista_interaction:
                        try:
                            clicker(driver, el)
                            elemento, html_elemento, _ = buscador(driver, tipo_busca, busca)
                            if elemento: break
                        except: continue

    except Exception as e:
        error_cod = type(e).__name__
        error_msg = str(e)
    finally:
        duracao = round(time.time() - inicio, 2)
        if driver: driver.quit()

    return html_selenium, elemento, html_elemento, error_cod, error_msg, texto_click, html_click, duracao
```

```python
def q_selenium_retry(df_to_retry, questao, tipo_busca, pv, pn, threads=4):
    urls = df_to_retry['link_avaliacao'].tolist()
    results = [None] * len(urls)

    with ThreadPoolExecutor(max_workers=threads) as executor:
        future_to_index = {executor.submit(scrap_retry, url, tipo_busca, pv, pn): idx for idx, url in enumerate(urls)}

        for future in as_completed(future_to_index):
            idx = future_to_index[future]
            try:
                results[idx] = future.result()
                print(f"Retry Progress: {idx + 1}/{len(urls)}")
            except Exception as exc:
                results[idx] = (None, None, None, "RETRY_THREAD_ERROR", str(exc), [], [], 0)

    res_df = pd.DataFrame(results, columns=['pagina_scraped', 'elemento_final', 'html_elemento_final', 'error_cod', 'error_msg', 'elemento_interacao_1', 'html_elemento_interacao_1', 'duracao_segundos'])
    return pd.concat([df_to_retry.reset_index(drop=True), res_df], axis=1)
```

```python
def q_selenium_paralelo(questao, tipo_busca, parametro_verificacao, parametro_navegacao, df=None, i=0, k=None, threads=4):
    # 1. Filtra os dados necessários
    qhtml = filtro_questao(questao, i, k, df)
    urls = qhtml['link_avaliacao'].tolist()

    # 2. Execução paralela
    results = [None] * len(urls)

    with ThreadPoolExecutor(max_workers=threads) as executor:
        # Mapeia a execução (scrap_selenium deve estar definida conforme as células anteriores)
        future_to_index = {executor.submit(scrap_selenium, url, tipo_busca, parametro_verificacao,parametro_navegacao ): idx for idx, url in enumerate(urls)}

        for future in as_completed(future_to_index):
            idx = future_to_index[future]
            try:
                data = future.result()
                results[idx] = data
                if (idx + 1) % 10 == 0:
                    print(f"Progresso: {idx + 1}/{len(urls)}")
            except Exception as exc:
                print(f"URL {urls[idx]} gerou uma exceção: {exc}")
                results[idx] = (None, None, None, "THREAD_ERROR", str(exc), [], [], 0)

    # 3. Organização dos resultados no DataFrame
    res_df = pd.DataFrame(results, columns=[
        'pagina_scraped', 'elemento_final', 'html_elemento_final',
        'error_cod', 'error_msg', 'elemento_interacao_1',
        'html_elemento_interacao_1', 'duracao_segundos'
    ])

    final_df = pd.concat([qhtml, res_df], axis=1)
    return final_df
```

```python
def q_selenium(questao, tipo_busca, parametro_verificacao, parametro_navegacao, df, i=0, k=None):

##Selecionando do DF a questão desejada

  qhtml = filtro_questao(questao, i, k, df)


##Iniciando o driver e aplicando o scraper por url

  results = []
  for idx, url in enumerate(qhtml['link_avaliacao'], start=1):
      results.append(scrap_selenium(url, tipo_busca, parametro_verificacao, parametro_navegacao)) #PARAMETRO VINDO DE DF QUESTAOxPARAMETROS
      if idx % 10 == 0:
          print(idx)


##Adicionando os resultados no DF
  results = pd.Series(results)
  qhtml['pagina_scraped'] = results.apply(lambda x: x[0])
  qhtml['elemento_final']        = results.apply(lambda x: x[1])
  qhtml['html_elemento_final']   = results.apply(lambda x: x[2])
  qhtml['error_cod']       = results.apply(lambda x: x[3])
  qhtml['error_msg']       = results.apply(lambda x: x[4])
  qhtml['elemento_interacao_1']       = results.apply(lambda x: x[5])
  qhtml['html_elemento_interacao_1']       = results.apply(lambda x: x[6])
  qhtml['duracao_segundos'] = results.apply(lambda x: x[7])

  return qhtml
```
