```python
def scrape_url(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        if pd.isna(url) or url == '':
            print("Erro: URL vazia ou não localizada para raspagem.")
            return False, "URL_EMPTY", "URL vazia ou não localizada para raspagem."
        else:
            response = requests.get(url, headers=headers, timeout=20)
            time.sleep(10)
            response.raise_for_status()
            time.sleep(10)
            soup = BeautifulSoup(response.content, 'html.parser')
            return soup, False, False

    except requests.exceptions.HTTPError as http_err:
        error_msg = f"HTTP error occurred for {url}: {http_err}"
        print(error_msg)
        error_cod = http_err.response.status_code
        error_msg = http_err.response.reason
        return False, error_cod, error_msg

    except requests.exceptions.ConnectionError as conn_err:
        error_msg = f"Connection error occurred for {url}: {conn_err}"
        print(error_msg)
        error_cod = type(conn_err).__name__
        error_msg = "ConnectionError"
        return False, error_cod, error_msg

    except requests.exceptions.Timeout as timeout_err:
        error_msg = f"Connection error occurred for {url}: {timeout_err}"
        print(error_msg)
        error_cod = type(timeout_err).__name__
        error_msg = "timeout_err"
        return False, error_cod, error_msg

    except requests.exceptions.RequestException as req_err:
        error_msg = f"An unexpected request error occurred for {url}: {req_err}"
        print(error_msg)
        error_cod = type(req_err).__name__
        error_msg = "req_err"
        return False, error_cod, error_msg

    except Exception as e:
        error_msg = f"An unexpected error occurred for {url}: {e}"
        print(error_msg)
        error_cod = e.response.status_code if hasattr(e, 'response') and hasattr(e.response, 'status_code') else type(e).__name__
        error_msg = e.response.reason if hasattr(e, 'response') and hasattr(e.response, 'reason') else str(e)
        return False, error_cod, error_msg
```

```python
def scrape_info_stealth(driver, url):
    if not url:
        print("Erro: URL vazia ou não localizada para raspagem.")
        return False, "URL_EMPTY", "URL vazia ou não localizada para raspagem."

    try:
        driver.get(url)
        driver.implicitly_wait(20)
        response_text = driver.page_source
        soup = BeautifulSoup(response_text, "html.parser")
        return soup, False, False

    except Exception as e:
        err = f"Erro ao raspar {url}: {e}"
        print(err)
        return False, type(e).__name__, str(e)
```

```python
def busca_todo_atributos(driver, element):
    #Retorna um dict com os atributos de cada elemento {nome_atributo: valor}
    js = """
        var items = {};
        for (var i = 0; i < arguments[0].attributes.length; i++) {
            items[arguments[0].attributes[i].name] = arguments[0].attributes[i].value;
        }
        return items;
    """
    return driver.execute_script(js, element)
```

```python
def localiza_clicks(driver, parametro_navegacao):
    # LOCALIZA SE HÁ ELEMENTOS COM TEXTOS SELECIONADOS
    lista_elementos = []
    vistos = set()

    for p in parametro_navegacao:
        try:
            els = driver.find_elements(By.XPATH, f'//*[contains(text(),"{p}")]') #BUSCA ELEMENTOS QUE TENHAM TEXTO P
            if not els:
              els = []

            elementos_atrib = driver.find_elements(By.XPATH, f"//*[@*[contains(.,'{p}')]]") #BUSCA ELEMENTOS QUE TENHAM ATRIBUTO P
            if not elementos_atrib:
              elementos_atrib = []

            for el in els + elementos_atrib:
                try:
                    txt = el.text
                    html = el.get_attribute("outerHTML")
                    id = (txt, html)

                    if id not in vistos and el not in lista_elementos:
                        lista_elementos.append(el)
                        vistos.add(id)
                except:
                    continue

        except NoSuchElementException:
            print('Elemento não localizado')
            pass

    return lista_elementos
```

```python
#Tenta realizar um click simples ou click JS  no elemento chamado
def clicker(driver,el):
  try:
    el.click()
    driver.implicitly_wait(5)

  except (ElementClickInterceptedException, ElementNotInteractableException):
    driver.execute_script('arguments[0].click();',el)
    driver.implicitly_wait(5)
```

```python
#retorna todo texto para os elementos desejados
def busca_todo_texto(driver, local):
    #LOCALIZACAO
    elements = driver.find_elements(By.CSS_SELECTOR, local)

    #lista de texto
    lista_text = []
    lista_html = []
    for e in elements:
      try:
        if e.text is not None:
          lista_text.append(e.text)
          lista_html.append(e.get_attribute("outerHTML"))

      except Exception as e:
        print(f"Erro ao obter atributos do elemento: {e}")
        pass

    lista_text = list(set(lista_text))
    lista_html = list(set(lista_html))
    return lista_text, lista_html
```

```python
#retorna link para uma parte de texto do link desejado
def busca_link(driver, parametro):
    #FUNCAO DE LOCALIZACAO
    #driver: driver seleneium ativo
    #parametro: lista de parametro string dos atributos que serão buscados, ex.: "radar", "horario"

    lista_elementos = [] #elementos que contenham algum p
    lista = []
    link = []
    html = []

    for p in parametro:
      elementos = driver.find_elements(By.XPATH, f"//*[@*[contains(.,'{p}')]]") #PARAMETRO DE BUSCA EM ATRIBUTOS

      if elementos is not None:
        lista_elementos.extend(elementos)

    for el in lista_elementos:
      if el.tag_name != 'input' and el.tag_name != 'option' and el.tag_name != 'form' and el.tag_name != "select" and el.get_attribute("option") == None:
        try:
          atribs = busca_todo_atributos(driver,el)

          for nome_atributo, valor_atributo in atribs.items():
            if (nome_atributo is not None) and (valor_atributo is not None) and any(p in valor_atributo for p in parametro):
              link.append(el.get_attribute(nome_atributo))
              html.append(el.get_attribute("outerHTML"))
              lista.append(el)

        except Exception as e:
          print(f"Erro ao obter atributos do elemento: {e}")

    lista = list(set(lista))
    link = list(set(link))
    html = list(set(html))

    return link, html, lista
```

```python
def buscador(driver, tipo_busca, busca):
    match tipo_busca:
      case "texto":
            elemento, html_elemento = busca_todo_texto(driver, busca)

      case "link":
            elemento, html_elemento, lista = busca_link(driver, busca)

    return elemento, html_elemento, lista
```

```python
def scrap_selenium(url, tipo_busca = None):

  inicio = time.time()
  busca = ["radardatransparencia.atricon"]        #PARAMETRO VINDO DE DF QUESTAOxPARAMETROS
  parametro_navegacao = padroniza_parametro(["Radar"]) #PARAMETRO VINDO DE DF QUESTAOxPARAMETROS
  html_selenium = None
  elemento = None
  html_elemento = None
  error_cod = False
  error_msg = False
  texto_click = []
  html_click = []
  elemento_f = None
  html_elemento_f = None
  duracao = 0

  if pd.isna(url) or (isinstance(url, str) and not url.strip()):
    print(f"Erro: URL inválida recebida: {repr(url)}")
    return None, None, None, "URL_INVALID", "URL vazia, None ou NaN.", None, None, None

  if ("https://radardatransparencia.atricon" in url) or ("https://radar.tce.mt." in url) or ("https://www.tce." in url):
    return None, None, None, "URL_VERIFY", "Foi verificado que a URL fornecida se refere diretamente ao Radar Transparencia ou TCE-MT", None, None, None

  driver = None

  try:
    driver = abrir_driver()
    driver.set_page_load_timeout(30)
    driver.get(url)
    wait = WebDriverWait(driver, 40)

    try: ##Parâmetros HTML para busca inicial
        wait.until(
            EC.any_of(EC.presence_of_all_elements_located((By.CSS_SELECTOR,".card-container")),
                      EC.presence_of_all_elements_located((By.TAG_NAME,"body")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".row")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".close-sub-itens")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".card")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".table")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".content")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".main-content")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".main-content-wrapper")),
                      EC.presence_of_all_elements_located((By.CSS_SELECTOR,".main"))
                      )
            )

        html_selenium = driver.page_source

    except TimeoutException as e:
        html_selenium = driver.page_source
        error_cod = type(e).__name__
        error_msg = str(e)
        print(f"Timeout esperando carregamento principal para {url}, seguindo mesmo assim.")

    try:
        url_inicial = driver.current_url
        driver.implicitly_wait(10)
        elemento, html_elemento, lista_elementos = buscador(driver,tipo_busca, busca)
        elemento, html_elemento, lista_elementos = buscador(driver,tipo_busca, busca)

        if not elemento:
          try:
            bt_aceito(driver)
            lista_elementos = localiza_clicks(driver, parametro_navegacao)
            if lista_elementos:
              for el in lista_elementos:
                texto_click.append(el.text)
                html_click.append(el.get_attribute("outerHTML"))

                clicker(driver,el)
                url_click = driver.current_url

                elemento, html_elemento, lista_elementos = buscador(driver,tipo_busca, busca)

                if not lista_elementos and url_click == url_inicial:
                  pass

                if not lista_elementos and url_click != url_inicial:
                  while url_click != url_inicial:
                    url_click = driver.current_url
                    driver.back()
                    driver.implicitly_wait(5)
                  pass

                if lista_elementos:
                  break

            if not lista_elementos:
              frases_bloqueio = ["Performing security verification", "Sorry, you have been blocked","403 Forbidden",
                                 "A página solicitada não foi encontrada no servidor.","The request could not be satisfied.",
                                 "Desculpe, não conseguimos encontrar o que você está procurando.",
                                 "This page is used to test the proper operation ","ERRO 500","Bad Gateway",
                                 "Blocked request this country"
                                ]
              bloqueio = verifica_bloqueio(driver, frases_bloqueio)

              if bloqueio == True:
                error_cod = "URL_BLOCKED"
                error_msg = "URL bloqueada"

              else:
                html_selenium = driver.page_source
                error_cod = "NENHUM_ELEMENTO_LOCALIZADO"
                error_msg = f"Nenhum elemento final localizado na página contendo: {busca} ou de interação contendo: {parametro_navegacao}"


          except NoSuchElementException as e:
              error_cod = "NoSuchElementException"
              error_msg = str(e)
              textos_corpo, html_corpo = busca_todo_texto(driver, "body")
              texto_click = textos_corpo
              html_click = html_corpo


    except Exception as e:
        if not error_cod:
            error_cod = type(e).__name__
            error_msg = str(e)
        print(f"Erro na busca de elementos para {url}: {e}")

  except WebDriverException as e:
      error_cod = type(e).__name__
      error_msg = str(e)
      print(f"WebDriver error for {url}: {error_msg}")

  except Exception as e:
      error_cod = type(e).__name__
      error_msg = str(e)
      print(f"General error for {url}: {error_msg}")

  finally:
    if (not elemento and not html_elemento) and (not error_msg and not error_cod):
      error_cod = "NENHUM_ELEMENTO_LOCALIZADO"
      error_msg = f"Nenhum elemento final localizado na página contendo: {busca} ou de interação contendo: {parametro_navegacao}"

    if (elemento and html_elemento) or (texto_click and html_click):
      error_cod = None
      error_msg = None

    duracao = round(time.time() - inicio, 2)

    if driver:
      print(f"Tempo de execução para {url}: {duracao}s")
      driver.quit()

  return html_selenium, elemento, html_elemento, error_cod, error_msg, texto_click, html_click, duracao
```

```python
def q_selenium(questao, i=0, k= None):

##Selecionando do DF a questão desejada

  qhtml = filtro_questao(questao, i, k)


##Iniciando o driver e aplicando o scraper por url

  results = []
  for idx, url in enumerate(qhtml['link_avaliacao'], start=1):
      results.append(scrap_selenium(url, "link")) #PARAMETRO VINDO DE DF QUESTAOxPARAMETROS
      if idx % 10 == 0:
          print(idx)


##Adicionando os resultados no DF
  results = pd.Series(results)
  qhtml['pagina_scraped'] = results.apply(lambda x: x[0])
  qhtml['elemento_final']        = results.apply(lambda x: x[1])
  qhtml['html_elemento_final']   = results.apply(lambda x: x[2])
  qhtml['error_cod']       = results.apply(lambda x: x[3])
  qhtml['error_msg']       = results.apply(lambda x: x[4])
  qhtml['elemento_interacao_1']       = results.apply(lambda x: x[5])
  qhtml['html_elemento_interacao_1']       = results.apply(lambda x: x[6])
  qhtml['duracao_segundos'] = results.apply(lambda x: x[7])

  return qhtml
```
